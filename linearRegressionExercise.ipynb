Exercise taken from [codebasics](https://github.com/codebasics/py/blob/master/ML/1_linear_reg/1_linear_regression.ipynb)

```
import pandas as pd
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```
import matplotlib.pyplot as plt
X = df.drop('per capita income (US$)', axis=1)
y = df['per capita income (US$)']
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
predictions = reg.predict(X)
plt.figure(figsize=(7, 4))
plt.scatter(X, y)
plt.plot(X, predictions)
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.show()
```

---
### Prediction canada's per capita income in year 2020:

```
reg.predict([[2020]])
```

```
reg.coef_*(2020) + reg.intercept_ # it's the same result as reg.predict([[2020]])
```
